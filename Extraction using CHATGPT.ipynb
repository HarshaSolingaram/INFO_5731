```python
import pandas as pd
import openai

# Set your API key here
openai.api_key = 'sk-proj-an01j2QUvV4cjgMoAe1ST3BlbkFJzrMBvUqN4G9Ik5JXKuBn'


# Function to extract information using ChatGPT
def extract_info_with_chatgpt(text):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{
                "role": "system",
                "content": " find me The purpose of the authors, The contribution of the paper, The method of the paper, The dataset used in the paper, The findings or conclusions of the paper, The future work of the paper"
            }, {
                "role": "user",
                "content": text
            }]
        )
        return response.choices[0].message['content'].strip()
    except Exception as e:
        return f"Error: {str(e)}"

# Apply the function to both dataframes
df['ExtractedOutput'] = df['TextData'].apply(extract_info_with_chatgpt)
newdf['ExpectedOutput'] = newdf['TextData'].apply(extract_info_with_chatgpt)

# Optionally, save the results to new CSV files
df.to_csv('processed_dataset.csv', index=False)
newdf.to_csv('processed_new_dataset.csv', index=False)
```
